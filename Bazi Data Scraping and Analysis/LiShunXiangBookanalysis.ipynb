```python
#https://stackoverflow.com/questions/28816046/
#displaying-different-images-with-actual-size-in-matplotlib-subplot
def display(im_path):
    dpi = 80
    im_data = plt.imread(im_path)

    height, width  = im_data.shape[:2]
    
    # What size does the figure need to be in inches to fit the image?
    figsize = width / float(dpi), height / float(dpi)

    # Create a figure of the right size with one axes that takes up the full figure
    fig = plt.figure(figsize=figsize)
    ax = fig.add_axes([0, 0, 1, 1])

    # Hide spines, ticks, etc.
    ax.axis('off')


    # Display the image.
    ax.imshow(im_data, cmap='gray')

    plt.show()
```

```python
# convert txts_2d to a string , each line is separated by a newline | and each word is separated by a space

def convert_txts_2d_to_str(txts_2d):
    txts_str = ''
    for line in txts_2d:
        for word in line:
            txts_str += word + ' '
        txts_str += '|'
    return txts_str
txts_str=convert_txts_2d_to_str(allresult[0:6])
txts_str
```

```python
from paddleocr import PaddleOCR,draw_ocr
def paddle(img_path):
    # Paddleocr supports Chinese, English, French, German, Korean and Japanese.
    # You can set the parameter `lang` as `ch`, `en`, `french`, `german`, `korean`, `japan`
    # to switch the language model in order.
    ocr = PaddleOCR(use_angle_cls=True, lang='ch') # need to run only once to download and load model into memory
    # img_path = 'host.jpg'
    # img_path = 'warped.jpg'
    # img_path = 'UH.tif'
    # img_path = img_paths[0]



    #show the image in new window
    # read the image
    # img = cv2.imread(img_path)
    # #show the image
    # cv2.imshow('image',img)


    # img_path = 'r.png'
    # img_path = 'claim_receipt.jpg'
    result = ocr.ocr(img_path, cls=True)
    for idx in range(len(result)):
        res = result[idx]
        # for line in res:
            # print(line)


    # draw result
    from PIL import Image
    result = result[0]
    image = Image.open(img_path).convert('RGB')
    boxes = [line[0] for line in result]
    txts = [line[1][0] for line in result]
    scores = [line[1][1] for line in result]
    y_up=[line[0][0][1] for line in result]
    y_down=[line[0][3][1] for line in result]
    x_left=[line[0][0][0] for line in result]
    x_right=[line[0][1][0] for line in result]
    im_show = draw_ocr(image, boxes, txts, scores, font_path='/path/to/PaddleOCR/doc/fonts/simfang.ttf')
    im_show = Image.fromarray(im_show)
    im_show.save('result.jpg')
    # im_show.show()
    display('result.jpg')
    return txts
```
