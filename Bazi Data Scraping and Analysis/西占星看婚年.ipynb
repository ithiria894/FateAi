```python
# X=MC
# T=火星
# O=海王星
# ‹=升交點
# P=冥王星
# E=水星
# R=金星
# U=土星
# Q=太陽
# W=月亮
# Z=AC
# I=天王星
# Y=木星

#convert the 事件 into corrasponding planet symbol
df.replace('X','MC',inplace=True)
df.replace('T','火',inplace=True)
df.replace('O','海王',inplace=True)
df.replace('‹','升交點',inplace=True)
df.replace('P','冥王',inplace=True)
df.replace('E','水',inplace=True)
df.replace('R','金',inplace=True)
df.replace('U','土',inplace=True)
df.replace('Q','日',inplace=True)
df.replace('W','月',inplace=True)
df.replace('Z','AC',inplace=True)
df.replace('I','天王',inplace=True)
df.replace('Y','木',inplace=True)
df.replace('<','福點',inplace=True)
```
