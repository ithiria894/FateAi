```python
from paddleocr import PaddleOCR
import fitz  # PyMuPDF
import os
from PIL import Image
import re

# Initialize the OCR model
ocr = PaddleOCR(use_angle_cls=True, lang='ch')  # need to run only once to download and load the model into memory

# PDF file path
pdf_path = r"c:\Users\TOSHIBA\Documents\大耕\紫微攻略1.pdf"

# Save extracted text to a file
output_txt_path = r"c:\Users\TOSHIBA\Documents\大耕\紫微攻略1.txt"

# Read the PDF file
doc = fitz.open(pdf_path)

# Initialize a variable to store all the text
all_text = ""

# Iterate through each page
for page_num in range(3):
    # Get the page
    page = doc.load_page(page_num)
    
    # Convert the page to an image
    pix = page.get_pixmap()
    
    # Save the image as a temporary file using Pillow
    img_path = f"temp_page_{page_num}.png"
    image = Image.frombytes("RGB", [pix.width, pix.height], pix.samples)
    image.save(img_path)
    
    # Use PaddleOCR to extract text from the image
    result = ocr.ocr(img_path, cls=True)
    
    # Extract and print Chinese characters only in reverse order
    chinese_text = ''
    for line in result:
        # Reverse the order of lines and words
        line = line[::-1]
        for word_info in line:
            text = word_info[1][0]
            # Concatenate Chinese characters
            chinese_text += ''.join(re.findall(r'[\u4e00-\u9fff]+', text))
            chinese_text += '\n'
    
    all_text += chinese_text
    
    # Add a newline character to separate pages
    all_text += '\n'
    all_text += '------------------------------------------------------------------------'
    all_text += '\n'

    # Remove the temporary image file
    os.remove(img_path)

# Save all the text to a file
with open(output_txt_path, "w", encoding="utf-8") as f:
    f.write(all_text)

print(f"Text has been extracted and saved to {output_txt_path}")
```
