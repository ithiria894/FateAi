```python
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
driver = webdriver.Chrome()
import pandas as pd
from selenium.webdriver.common.by import By

def getDayun(year,month,day,year_of_marriage,index):
    
    year=str(year)
    month=str(month)
    day=str(day)
    hr="0"
    gender="Male"

    driver.get("https://fatew.com/")
    select = Select(driver.find_element(By.NAME, "hityear"))
    select.select_by_value(year)

    select = Select(driver.find_element(By.NAME, "hitmonth"))
    select.select_by_visible_text(month)

    select = Select(driver.find_element(By.NAME, "hitday"))
    select.select_by_visible_text(day)

    select = Select(driver.find_element(By.NAME, "hittime"))
    # select.select_by_value(hr)
    select.select_by_visible_text(hr)

    if gender == "Male":
        driver.find_elements(By.ID,'male')[0].click()
        #click on the radio button by class="myradio"
        # driver.find_element_by_css_selector("input[type='radio'][value='1']").click()
        # modify he above line to make it work, there s AttributeError: 'WebDriver' object has no attribute 'find_element_by_css_selector'
        

    if gender =="Female":
        driver.find_elements(By.ID,'female')[0].click()

    driver.find_elements(By.CSS_SELECTOR,"input[type='submit'][id='submitb']")[0].click()

    returnv=driver.find_elements(By.XPATH,"//font[@size = '6' and @face='標楷體']")

    baz=""
    for i in returnv:
        baz=baz+i.text

    baz=baz.replace("\n", "")
    b=baz[0:4]+baz[5:]
    bazi=list(b.strip())
    print(bazi)



    table_element = driver.find_elements(By.XPATH,'//table[@border="1" and @cellspacing="0" and @cellpadding="0" and @bordercolor="#F2F2F2"]')

    # Convert the table element into a string of HTML
    table_html = table_element[0].get_attribute('outerHTML')

    # Use Pandas to parse the HTML string into a DataFrame
    df = pd.read_html(table_html)[0]

    #make first row as header
    new_header = df.iloc[0] #grab the first row for the header
    df = df[1:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    #rename df to dayunliunian
    dayunliunian=df
    #save to csv
    dayunliunian.to_csv(f"{index}dayun.csv",index=False)
    # if year_of_marriage == 西元 in dayunliunian: get the carrasponding 大運 at that year
    dayun=dayunliunian[dayunliunian['西元'] == str(year_of_marriage)]['大運'].values[0]

    return dayun

```
