```python
def solution(inputString):
    p1=0
    p2=len(inputString)-1
    
    while(p1!=p2):
        if inputString[p1]!=inputString[p2]:
            return False
        p1=p1+1
        p2=p2-1
    return True
solution('hlbeeykoqqqqokyeeblh')
```
