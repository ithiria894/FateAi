```python
#get only  Registration_Date Groom_Birth_Date Age Groom_Income Groom_Occupation Marriage_Date Marriage_Years Number_of_Children
selected_columns = ['Registration_Date', 'Groom_Birth_Date', 'Age', 'Groom_Income', 'Groom_Occupation',
                    'Marriage_Date', 'Marriage_Years', 'Number_of_Children']

man_df_selected = man_df[selected_columns]
```

```python
#drop if Groom_Birth_Date is NaN
man_df_selected = man_df_selected.dropna(subset=['Groom_Birth_Date'])

#convert Registration_Date Groom_Birth_Date Marriage_Date to datetime

man_df_selected['Registration_Date'] = pd.to_datetime(man_df_selected['Registration_Date'])
man_df_selected['Groom_Birth_Date'] = pd.to_datetime(man_df_selected['Groom_Birth_Date'])
man_df_selected['Marriage_Date'] = pd.to_datetime(man_df_selected['Marriage_Date'])
```

```python
# create col Age_at_Marriage 
man_df_selected['Age_at_Marriage'] = (man_df_selected['Marriage_Date'] - man_df_selected['Groom_Birth_Date']).dt.days // 365
```
