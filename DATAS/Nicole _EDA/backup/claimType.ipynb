```python
# remove the None value from the list
# rscd_des_lst.remove(None)

# make all string to upper case
rscd_des_lst = [rscd_des.upper() for rscd_des in rscd_des_lst]
# get only medical hospital clinical related rscd_des from the list
medical_rscd_des_lst = [rscd_des for rscd_des in rscd_des_lst if "MEDICAL" in rscd_des or "HOSPITAL" in rscd_des or "CLINICAL" in rscd_des]
```

```python
# get all rows where rscd_des is in the medical_rscd_des_lst
medical_rscd_des_df = clambase_df[clambase_df["rscd_des"].isin(medical_rscd_des_lst)]
medical_rscd_des_df
```

```python
# get only important columns
useful_clambase_df = medical_rscd_des_df[["DATOCC","DTEEFF","DATREP","rscd_des","contrnb"]]
```

```python
CLIENT_DATA_MAPPING_FILE = os.path.join(PROCESSED_DATA_DIR, "processed_client_data.parquet")
# read in processed_client_data_df
client_data_df = pd.read_parquet(CLIENT_DATA_MAPPING_FILE)
```

```python
client_data_df
```

```python
#get only important columns from client_data_df
useful_client_data_df = client_data_df[["CLNTNUM","CLTSEX","MARRYD","CLTDOB"]]
```

```python
# drop duplicate rows of client_data_df by all columns
useful_client_data_df = useful_client_data_df.drop_duplicates()
```

```python
#merge useful_client_data_df with p_c on COWNNUM and CLNTNUM on the right
p_c_c = pd.merge(p_c, useful_client_data_df, left_on='COWNNUM', right_on='CLNTNUM', how='left')
```

```python
p_c_c
```

```python
# merge CLTDOB_x and CLTDOB_y to CLTDOB, get the cell with non NaN value
p_c_c["CLTDOB"] = p_c_c["CLTDOB_x"].combine_first(p_c_c["CLTDOB_y"])
```

```python
# combine CLTSEX_x and CLTSEX_y to CLTSEX, get the cell with non NaN value
p_c_c["CLTSEX"] = p_c_c["CLTSEX_x"].combine_first(p_c_c["CLTSEX_y"])
```

```python
# combine MARRYD_x and MARRYD_y to MARRYD, get the cell with non NaN value
p_c_c["MARRYD"] = p_c_c["MARRYD_x"].combine_first(p_c_c["MARRYD_y"])
```

```python
# drop CLTDOB_x, CLTDOB_y, CLTSEX_x, CLTSEX_y, MARRYD_x, MARRYD_y
p_c_c = p_c_c.drop(columns=["CLTDOB_x","CLTDOB_y","CLTSEX_x","CLTSEX_y","MARRYD_x","MARRYD_y"])
```

```python
p_c_c
```

```python
#drop if CLTDOB is NaN
p_c_c = p_c_c.dropna(subset=["CLTDOB"])
```

```python
p_c_c
```

```python
# drop if CLTDOB is 99999999.0
p_c_c = p_c_c[p_c_c["CLTDOB"] != 99999999.0]
```

```python
p_c_c
```

```python
# drop is CLTSEX is NaN
p_c_c = p_c_c.dropna(subset=["CLTSEX"])
```

```python
p_c_c
```

```python
#conver DATOCC DTEEFF DATREP to datetime
p_c_c["DATOCC"] = pd.to_datetime(p_c_c["DATOCC"], format="%Y%m%d")
p_c_c["DTEEFF"] = pd.to_datetime(p_c_c["DTEEFF"], format="%Y%m%d")
p_c_c["DATREP"] = pd.to_datetime(p_c_c["DATREP"], format="%Y%m%d")
```

```python
# create a new col called DATOCC_year and get the year of DATOCC
p_c_c["DATOCC_year"] = p_c_c["DATOCC"].dt.year
# create a new col called DATOCC_month and get the month of DATOCC
p_c_c["DATOCC_month"] = p_c_c["DATOCC"].dt.month
```

```python
# conver CLTDOB to datetime
p_c_c["CLTDOB"] = pd.to_datetime(p_c_c["CLTDOB"], format="%Y%m%d")
```

```python
# get only age DATOCC DATOCC_year DATOCC_month rscd_des CLTDOB CLTSEX MARRYD and output to parquet
p_c_c_out = p_c_c[["DATOCC","DATOCC_year","DATOCC_month","rscd_des","CLTDOB","CLTSEX","MARRYD"]]
```
