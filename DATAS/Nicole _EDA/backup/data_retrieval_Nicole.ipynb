# One off, convert byte string to string and save as parquet

## convert client id mapping file

```python
CLIENT_SAS_FILE = os.path.join(GI_DATA_DIR, "client.sas7bdat")

# import client file

client_df = pd.read_sas(CLIENT_SAS_FILE)

# select str column

str_df = client_df.select_dtypes([object])

# convert str df

str_df = str_df.stack().str.decode('utf-8').unstack()

# replace original df

for col in str_df:
    client_df[col] = str_df[col].fillna(client_df[col])

# save as parquet

client_df.to_parquet(os.path.join(GI_DATA_DIR, "client.parquet"))

# free memroy

del client_df

gc.collect()
```
