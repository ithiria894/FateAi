```python
def minStartValue(self, nums):
    prefix_sum = 0 
    min_start_value = 1  
    for num in nums: 
        prefix_sum += num             
    min_start_value = max(min_start_value, 1-prefix_sum)         
    return min_start_value
```

```python
def find_max_num(arr, num):
    max_num = num  # Initialize max_num with the given value of num

    for element in arr:
        if element == max_num:
            max_num *= 3

    return max_num
```
