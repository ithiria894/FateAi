```python
from datetime import datetime

def solution(schedule, time):
    # ft=[]
    lastBus=-1
    # change each time in schedule to datetime
    # change time to datetime
    time=datetime.strptime(time, '%H:%M')
    for t in schedule:
        t=datetime.strptime(t, '%H:%M')
        if time> t:
            lastBus=t
    print(lastBus)
    print(time-lastBus)
    diff=time-lastBus
    # return only the minutes
    return(diff.seconds//60)
```

```python
def solution(matrix):
    sharp=[]
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == '#':
                sharp.append([i,j])
    print(sharp)
    distance=[]
    for s in sharp:
        print("-")
        for i in reversed(range(0,s[0])):
            print(i)
            # check if matrix[i][s[1]] is F
            if matrix[i][s[1]] == 'F':
                print("found F")
                print("distance:", s[0]-i-1)
                distance.append(s[0]-i-1)
    minDistance=min(distance)
    # move the whole matrix down by minDistance keep the # in the same place
    #create a list with . with lenth of len(matrix[i])
    # add the list to the top of the matrix
    # remove the last list from the matrix
    for i in range(minDistance):
        matrix.insert(0,['.' for i in range(len(matrix[0]))])
        matrix.pop()
        
    # replace all # with .
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == '#':
                matrix[i][j]='.'
        
    # put # back in the same place
    for s in sharp:
        matrix[s[0]][s[1]]='#'
        
    return matrix
    
        
```
