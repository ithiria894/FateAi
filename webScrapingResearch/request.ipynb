```python
import requests
from bs4 import BeautifulSoup

year="1986"
month="6"
day="10"
hr="14"
minute="30"
gender="F"
# Define the URL for each website
urls = {
    'https://fatew.com/star/': {
        'params': {
            'hityear': year,
            'hitmonth': month,
            'hitday': day,
            'hittime': hr,
            'sex': gender
        }
    },
    'https://www.dearmoney.com.tw/eightwords': None,
    'https://myfate.herokuapp.com/': {
        'params': {
            'Year': year,
            'Month': month,
            'Day': day,
            'Hour': hr,
            'Sex': gender
        }
    },
    'https://mingming3.com/home/life': {
        'params': {
            'o_year': year,
            'o_month': month,
            'o_day': day,
            'o_hour': hr,
            'gender': 'm' if gender == 'M' else 'f'
        }
    },
    'https://astrodoor.cc/horoscope.jsp': {
        'params': {
            'year': year,
            'month': month,
            'day': day,
            'hour': hr,
            'minute': minute,
            'province': '1'
        }
    }
}

# Iterate over the URLs and make requests
for url, params in urls.items():
    response = requests.get(url, params=params)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Process the HTML content as needed
    if url == 'https://mingming3.com/home/life':
        # Example processing for the specific website
        divs = soup.find_all('div', class_='content-container')
        for div in divs:
            print('---------------------------------------')
            print(div.text)
    elif url == 'https://astrodoor.cc/horoscope.jsp':
        # Example processing for another website
        # You can add similar processing for other websites as needed
        pass
    else:
        # Default processing
        print(soup.prettify())  # Print the HTML content
```
