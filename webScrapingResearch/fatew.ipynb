```python
import requests
from bs4 import BeautifulSoup

year = "1986"
month = "6"
day = "10"
hour = "14"
minute = "30"
gender = "F"

url = "https://fatew.com/star/"
params = {
    "hityear": year,
    "hitmonth": month,
    "hitday": day,
    "hittime": hour,
    "hitminute": minute,
    "gender": gender
}

response = requests.get(url, params=params)

if response.status_code == 200:
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find and print all text content
    for element in soup.find_all(string=True):
        print(element.strip())
        # //print all content in the webpage

        
else:
    print("Failed to fetch data from the website.")
```

```python
import requests
from bs4 import BeautifulSoup

# Define the URL of the website
url = "https://fatew.com/star/"

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the response
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the select element with the name "hityear"
    select_element = soup.find("select", {"name": "hityear"})
    
    # Check if the select element was found
    if select_element:
        # Find all option elements within the select element
        option_elements = select_element.find_all("option")
        
        # Loop through each option element
        for option in option_elements:
            # Print the value of each option
            print(option["value"])
else:
    # Print an error message if the request was not successful
    print("Failed to fetch data from the website.")
```
